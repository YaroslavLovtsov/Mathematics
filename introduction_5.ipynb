```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = figure()
ax1 = Axes3D(fig)

X = np.arange(-7,7,0.33)
Y = np.arange(-7,7,0.33)

X,Y = meshgrid(X,Y)

Z1 = 3/5*X +4/5*Y + 2
Z2 = 3/5*X +4/5*Y - 7

ax1.plot_wireframe(X,Y,Z1)
ax1.plot_wireframe(X,Y,Z2)
show()
```

```python
fig = figure()
ax2 = Axes3D(fig)

X = np.arange(-7,7,0.33)
Y = np.arange(-6,6,0.33)

X,Y = meshgrid(X,Y)

Z1 = 15*X*X - 20*Y*Y -7
Z2 = 200 + 16*X*X + 9*Y*Y

ax2.plot_wireframe(X,Y,Z1)
ax2.plot_wireframe(X,Y,Z2)
show()
```
