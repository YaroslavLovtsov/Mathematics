```python
%matplotlib inline
import numpy as np 
import matplotlib.pyplot as plt

def get_cartesius_point(polar_point):
    return [polar_point[0] * np.cos(polar_point[1]),polar_point[0] * np.sin(polar_point[1])]


test_polar_point = [2,1]
print(get_cartesius_point(test_polar_point))
```

```python
phi = np.linspace(1.15,4.2,1000)

plt.polar(phi, 3 / (np.sin(phi) - 2*np.cos(phi)))
```

```python
psi = np.linspace(0, 2 * np.pi,300)
plt.polar(psi,3+psi-psi)
```
