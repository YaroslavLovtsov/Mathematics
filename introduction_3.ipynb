```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-4,4,128)

print("Окружность")

plt.plot(x,np.sqrt(16 - x**2), marker = ".")
plt.plot(x,-np.sqrt(16 - x**2), marker = ".")
```

```python
x = np.linspace(-4,4,128)

print("Эллипс")

plt.plot(x,np.sqrt(9/16 *(16 -  x**2)), marker = ".")
plt.plot(x,-np.sqrt(9/16 * (16 - x**2)), marker = ".")
```

```python
x = np.linspace(-7,7,128)

print("Гипербола")

plt.plot(x,np.sqrt(1/16 *(16 +  x**2)), marker = ".")
plt.plot(x,-np.sqrt(1/16 * (16 + x**2)), marker = ".")
```
